```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0.01, 20)
```

```python
figure, axis = plt.subplots(2)
y = np.log(x)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
figure, axis = plt.subplots(2)
y = x * np.log2(x)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
figure, axis = plt.subplots(2)
y = x * np.log10(x)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
x = np.linspace(-20, 20)
```

```python
figure, axis = plt.subplots(2)
y = np.exp(x)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
figure, axis = plt.subplots(2)
y = np.power(x, 5)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
figure, axis = plt.subplots(2)
y = 1/np.power(x, 5)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
figure, axis = plt.subplots(2)
y = np.abs(x)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
figure, axis = plt.subplots(2)
y = np.power(x, 2) + np.power(x, 4)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```

```python
figure, axis = plt.subplots(2)
y = np.power(x, 4) - np.power(x, 2)
axis[0].plot(x, y)
axis[1].plot(x, -y)
```
